```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import pandas as pd
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import os

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

if not os.path.exists("PEMS-BAY.csv"):
    print("Downloading PEMS-BAY dataset...")
    !wget -O PEMS-BAY.csv "https://zenodo.org/records/5146275/files/PEMS-BAY.csv?download=1"

if not os.path.exists("METR-LA.csv"):
    print("Downloading METR-LA dataset...")
    !wget -O METR-LA.csv "https://zenodo.org/records/5146275/files/METR-LA.csv?download=1"

def get_num_nodes(csv_file):
    data = pd.read_csv(csv_file).select_dtypes(include=[np.number])
    return data.shape[1]

def preprocess_csv_inductive(filepath, seq_length=12, pred_length=12, sample_frac=1.0, seen_frac=0.1):
    traffic_data = pd.read_csv(filepath)
    traffic_data = traffic_data.select_dtypes(include=[np.number]).values

    if sample_frac < 1.0:
        num_samples = int(len(traffic_data) * sample_frac)
        traffic_data = traffic_data[:num_samples]

    mean = traffic_data.mean()
    std = traffic_data.std()
    traffic_data = (traffic_data - mean) / std

    num_nodes = traffic_data.shape[1]
    num_seen = int(num_nodes * seen_frac)

    seen_nodes = np.random.choice(num_nodes, num_seen, replace=False)
    unseen_nodes = np.setdiff1d(np.arange(num_nodes), seen_nodes)

    def create_sequences(data, nodes):
        X, Y = [], []
        for i in range(len(data) - seq_length - pred_length):
            X.append(data[i:i + seq_length, nodes])
            Y.append(data[i + seq_length:i + seq_length + pred_length, nodes])
        return np.array(X), np.array(Y)

    X_seen, Y_seen = create_sequences(traffic_data, seen_nodes)
    X_unseen, Y_unseen = create_sequences(traffic_data, unseen_nodes)

    adjacency_matrix = np.corrcoef(traffic_data.T)
    adjacency_matrix = adjacency_matrix[np.ix_(seen_nodes, seen_nodes)]
    adjacency_matrix[np.isnan(adjacency_matrix)] = 0

    unseen_adjacency_matrix = np.corrcoef(traffic_data.T)
    unseen_adjacency_matrix = unseen_adjacency_matrix[np.ix_(unseen_nodes, unseen_nodes)]
    unseen_adjacency_matrix[np.isnan(unseen_adjacency_matrix)] = 0

    return (
        DataLoader(TensorDataset(torch.tensor(X_seen, dtype=torch.float32),
                                 torch.tensor(Y_seen, dtype=torch.float32)),
                   batch_size=16, shuffle=True),
        DataLoader(TensorDataset(torch.tensor(X_unseen, dtype=torch.float32),
                                 torch.tensor(Y_unseen, dtype=torch.float32)),
                   batch_size=16, shuffle=False),
        adjacency_matrix,
        unseen_adjacency_matrix,
        len(seen_nodes),
        len(unseen_nodes)
    )

class CNFGNN(nn.Module):
    def __init__(self, adjacency_matrix, num_nodes, hidden_size=64):
        super().__init__()
        self.adjacency = nn.Parameter(torch.tensor(adjacency_matrix, dtype=torch.float32), requires_grad=False)
        self.fc1 = nn.Linear(num_nodes, hidden_size)
        self.fc2 = nn.Linear(hidden_size, num_nodes)

    def forward(self, x):
        batch_size, seq_length, num_nodes = x.size()

        if self.adjacency.shape[0] != num_nodes:
            raise ValueError(f"Adjacency matrix size {self.adjacency.shape} does not match input size {num_nodes}.")

        x = torch.einsum('ij,btj->bti', self.adjacency, x)
        x = x.mean(dim=1)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)

        return x

def evaluate_rmse(model, test_loader):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for X_batch, Y_batch in test_loader:
            X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
            output = model(X_batch)
            target = Y_batch.mean(dim=1)
            y_true.extend(target.cpu().numpy().flatten())
            y_pred.extend(output.cpu().numpy().flatten())
    return np.sqrt(mean_squared_error(y_true, y_pred))

def train_federated_cnfgnn(adjacency_matrix, seen_loader, unseen_loader, num_seen, num_unseen, unseen_adjacency_matrix, epochs=10):
    local_models = [CNFGNN(adjacency_matrix, num_seen, hidden_size=64).to(device) for _ in range(5)]
    global_model = CNFGNN(adjacency_matrix, num_seen, hidden_size=64).to(device)

    optimizer = torch.optim.Adam(global_model.parameters(), lr=0.001)
    criterion = nn.MSELoss()

    test_rmse_values = []

    for epoch in range(epochs):
        global_state = global_model.state_dict()
        local_states = []
        for local_model in local_models:
            local_model.load_state_dict(global_state)
            local_optimizer = torch.optim.Adam(local_model.parameters(), lr=0.001)

            for X_batch, Y_batch in seen_loader:
                X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
                local_optimizer.zero_grad()
                output = local_model(X_batch)
                target = Y_batch.mean(dim=1)
                loss = criterion(output, target)
                loss.backward()
                local_optimizer.step()

            local_states.append(local_model.state_dict())

        global_state = {key: torch.stack([local_state[key] for local_state in local_states], dim=0).mean(dim=0) for key in global_state.keys()}
        global_model.load_state_dict(global_state)

        unseen_model = CNFGNN(unseen_adjacency_matrix, num_unseen, hidden_size=64).to(device)

        rmse = evaluate_rmse(unseen_model, unseen_loader)
        test_rmse_values.append(rmse)
        print(f"Epoch {epoch + 1}: Test RMSE on Unseen Nodes = {rmse:.4f}")

    return test_rmse_values

if __name__ == "__main__":
    datasets = ["PEMS-BAY.csv", "METR-LA.csv"]
    seen_fractions = [0.05, 0.1, 0.2, 0.25, 0.5, 0.75]

    for dataset in datasets:
        print(f"\nEvaluating CNFGNN in Inductive Setting on {dataset}...")

        for seen_frac in seen_fractions:
            print(f"\nTraining with {int(seen_frac * 100)}% seen nodes...")

            seen_loader, unseen_loader, adjacency_matrix, unseen_adjacency_matrix, num_seen, num_unseen = preprocess_csv_inductive(dataset, seen_frac=seen_frac)

            test_rmse_values = train_federated_cnfgnn(
                adjacency_matrix,
                seen_loader,
                unseen_loader,
                num_seen,
                num_unseen,
                unseen_adjacency_matrix,
                epochs=10
            )

            plt.figure()
            plt.plot(range(1, len(test_rmse_values) + 1), test_rmse_values, label=f"{int(seen_frac * 100)}% Seen Nodes")
            plt.title(f"Test RMSE for {dataset} ({int(seen_frac * 100)}% Seen Nodes)")
            plt.xlabel("Epoch")
            plt.ylabel("RMSE")
            plt.legend()
            plt.grid()
            plt.show()
```
